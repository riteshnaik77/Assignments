```python
#Lets normalize tge data for a better understanding of plot
```

```python
def norm_func(i):
       X = (i-i.min())/(i.max()-i.min())
       return (X)
```

```python
df_norm = norm_func(df.iloc[:,1:])
```

```python
df_norm
```

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch
```

```python
type(df_norm)
```

```python
z = linkage(df_norm, method="complete",metric="euclidean")
```

```python
z
```

```python
plt.figure(figsize=(15, 5));plt.title('Hierarchical Clustering Dendrogram');plt.xlabel('Index');plt.ylabel('Distance')
sch.dendrogram(z,leaf_rotation=0.,leaf_font_size=8.)
plt.show()
```

```python
# Lets apply Agglomerative Clustering. choosing 3 as clusters from the dendrogram
```

```python
from sklearn.cluster import AgglomerativeClustering
h_complete=AgglomerativeClustering(n_clusters=3,linkage='complete',affinity = "euclidean").fit(df_norm) 

cluster_labels=pd.Series(h_complete.labels_)

df['clust']=cluster_labels # creating a  new column and assigning it to new column 
```

```python
df = df.iloc[:,[5,0,1,2,3,4]]
df.head()
```

```python
# getting aggregate mean of each cluster
df.iloc[:,2:].groupby(df.clust).median()
```
