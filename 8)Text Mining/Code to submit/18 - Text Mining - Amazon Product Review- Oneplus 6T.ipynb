```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
link = 'https://www.amazon.in/OnePlus-Mirror-Black-128GB-Storage/product-reviews/B07DJHV6VZ/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
```

```python
page = requests.get(link)
```

```python
page
```

```python
page.content
```

```python
soup = bs(page.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
names = soup.find_all('span',class_='a-profile-name')
```

```python
names
```

```python
cust_name = []
for i in range(0,len(names)):
    cust_name.append(names[i].get_text())
cust_name
```

```python
cust_name.pop(0)
```

```python
cust_name.pop(0)
```

```python
cust_name
```

```python
title = soup.find_all('a',class_='review-title-content')
```

```python
title
```

```python
review_title = []
for i in range(0,len(title)):
    review_title.append(title[i].get_text())
review_title
```

```python
review_title[:] = [titles.lstrip('\n') for titles in review_title]
review_title
```

```python
review_title[:] = [titles.rstrip('\n') for titles in review_title]
review_title
```

```python
rating = soup.find_all('i',class_='review-rating')
rating
```

```python
rate = []
for i in range(0,len(rating)):
    rate.append(rating[i].get_text())
rate
```

```python
rate.pop(0)
```

```python
rate.pop(0)
```

```python
rate
```

```python
review = soup.find_all("span",{"data-hook":"review-body"})
review
```

```python
review_content = []
for i in range(0,len(review)):
    review_content.append(review[i].get_text())
review_content
```

```python
review_content[:] = [reviews.lstrip('\n') for reviews in review_content]
review_content
```

```python
review_content[:] = [reviews.rstrip('\n') for reviews in review_content]
review_content
```

```python
len(review_content)
```

```python
cust_name
review_title
rate
review_content
```

```python
df = pd.DataFrame()
```

```python
df
```

```python
df['Customer Name']=cust_name
```

```python
df
```

```python
df['Review title']=review_title
df['Ratings']=rate
df['Reviews']=review_content
```

```python
df
```

```python
df.to_csv(r'E:\reviews.csv',index=True)
```

```python
#Lets run sentiment analysis on the same dataset
```

```python
################## Sentiment Analysis #####################3
from textblob import TextBlob
df['polarity'] = df['Reviews'].apply(lambda x: TextBlob(x).sentiment[0])
df[['Reviews','polarity']].head(5)

# Displaying top 5 positive posts of Category_A
df[df.polarity>0].head(5)
```

```python
# ======= The distribution of Categories polarity score =======
import seaborn as sns  
import matplotlib.pyplot as plt 
sns.set()
plt.hist(x='polarity', data=df, bins=20);
plt.xlabel('polarity of Reviews');
plt.ylabel('count'); 
plt.figsize=(10, 16)
```
